```python
import pandas as pd
import regex as re
```

```python
emoticonDictionary = {
    'happy': ['(:', ':-)', ':)', ':-]', ':]', ':->', ':>', '8-)', '8)', ':-}', ':}', ':o)', ':c)',
              ':^)', '=]', '=)', ':happy:', ':smile:', ':welcome:', ':счастливый:', ':yeah:', ':счастлив:'],
    'laugh': [':-D', ':D', '8-D', '8D', '=D', '=3', 'B^D', 'c:', 'C:', 'x-D', 'xD', 'X-D', 'XD', '^.^', '^_^.',
              ':laughing:', ':смеется:', ':смех:', ':смеясь:', ':смеющийся:', ':joyous:', ':laughin:', ':rolling:'], # '^^'
    'sad': [':-(', ':(', ':-c', ':c', ':-<', ':<', ':-[', ':[', ':-||', ':{', ':@', ':(', ';(', '-_-', ':frustrating:'],
    'cry': [':\'-(', ':\'(', ':=(', ';-;', ':dry:'],
    'angry': ['>:(', '>:[', ':angry:', ':redface:', ':покраснел:'],
    'surprise': [':-O', ':O', ':-o', ':o', ':-0', ':0', '8-0', '>:O', '=O', '=o', '=0', 'O_O', 'o_o', 'O-O', 'o-o', 'O_o',
                 'o_O', '0.0', '*_*', '@_@', '@.@'],
    'catface': [':-3', ':3', '=3', 'x3', 'X3', '>:3', '=^_^=', ':kitteh:', ':kitten:', ':lovekitty:', ':hellokitty:'],
    'kiss': [':-*', ':*', ':×'],
    'wink': [';-)', ';)', '*-)', '*)', ';-]', ';]', ';^)', ';>', ':-,', ';D', ';3', ':wink:', ':подмигивание:', ':подмигнуть:'],
    'tongueout': [':-P', ':P', 'X-P', 'XP', 'x-p', 'xp', ':-p', ':p', ':-Þ', ':Þ', ':-þ', ':þ', ':-b', ':b', 'd:', '=p',
                  '>:P', ':tongue:', ':язык:'],
    'undecided': [':-/', ':/', ':-.', '>:\\\\', '>:/', ':\\\\', '=/', '=\\\\', ':L', '=L', ':S', ':unsure:'],
    'noexpression': [':-|', ':|'],
    'embarrassed': [':$', '://3', ':mellow:'],
    'halo': ['O:-)', 'O:)', '0:-3', '0:3', '0:-)', '0:)', '0;^)'],
    'evil': ['>:-)', '>:)', '}:-)', '}:)', '3:-)', '3:)', '>;-)', '>;)', '>:3', ';3', ':devilish:'],
    'confused': ['%)', '?_?', ':шокирован:'],
    'brokenheart': ['</3', '<\\3'],
    'heart': ['<3', '❤️', '♡', '❤️', ':hearts:', '♥️', ':hearteyes:'],
    'fainted': ['X_X', 'x_x', '+_+', 'X_x', 'x_X', ':exterminate:'],
    'lookleft': ['<_<', '<.<'],
    'lookright': ['>_>', '>.>'],
    'skeptical': ['>.<'],
    'troubled': ['>_<', '>w<'],
    'nervous': ['^_^;', '-_-;', '~_~', '^^;', ':eek:'],
    'shame': ['-.-'],
    'tired': ['=_=', ':скучно:'],
    'brows': ['>:'],
    'rolleyes': [':rolleyes:', ':rolldeyes:'],
    'proud': [':гордый:', ':гордость:'],
    'hug': [':hug:', ':hugs:'],
    'madness': [':сумасшедший:', ':псих:']
}

emoticonDictionaryRegex = {
    'sad': [';_+;'],
    'happy': [':-\)+', ':\)+'],
    'sadness': ['\._+\.'],
    'complexface': [' ͡° ͜ʖ ͡°', 'ಠ_+ಠ', 'ヽ༼ຈل͜ຈ༽ﾉ', '◕‿◕', '¯\\\\_(ツ)_/¯', '¯_\(ツ\)_/¯'],
    'joyful': ['\^_+\^', '\^-\^'],
    'arrowleft': ['<-+', '<=+', '<{3,}'],
    'arrowright': ['-+>', '=+>', '>{3,}']
}

for key, values in emoticonDictionary.items ():
    lststr = list (values)
    for i in range (0, len (lststr)):
        lststr [i] = re.escape (lststr [i])
    emoticonDictionary [key] = lststr

def replaceEmoticons (post):
    for key, values in emoticonDictionaryRegex.items ():
        for item in values:
            post = re.sub (item, ' emoticon ' + key + ' ', post)
    for key, values in emoticonDictionary.items ():
        for item in values:
            post = re.sub (item, ' emoticon ' + key + ' ', post)
    return post
```

```python
data1 = pd.read_csv ('without_url.csv')
data1.head ()
```

```python
def filterAll (post):
    post = re.sub ('\\\\"', '\"', post)
    if ((post [0] == '\'') or (post [0] == '\"') or (post [0] == '«')):
        post = ' ' + post [1: ]
    else:
        post = ' ' + post
    if ((post [- 1] == '\'') or (post [- 1] == '»')):
        post = post [: - 1] + ' '
    
    post = re.sub ('«', '\"', post)
    post = re.sub ('»', '\"', post)
    
    post = re.sub ('(?:E|e|Е|е)(?:N|n|Н|н)(?:F|f|Ф|ф)(?:P|p|П|п)', ' ENFP ', post)
    post = re.sub ('(?:E|e|Е|е)(?:N|n|Н|н)(?:F|f|Ф|ф)(?:J|j)', ' ENFJ ', post)
    post = re.sub ('(?:E|e|Е|е)(?:N|n|Н|н)(?:T|t|Т|т)(?:P|p|П|п)', ' ENTP ', post)
    post = re.sub ('(?:E|e|Е|е)(?:N|n|Н|н)(?:T|t|Т|т)(?:J|j)', ' ENTJ ', post)
    post = re.sub ('(?:E|e|Е|е)(?:S|s|С|с)(?:F|f|Ф|ф)(?:P|p|П|п)', ' ESFP ', post)
    post = re.sub ('(?:E|e|Е|е)(?:S|s|С|с)(?:F|f|Ф|ф)(?:J|j)', ' ESFJ ', post)
    post = re.sub ('(?:E|e|Е|е)(?:S|s|С|с)(?:T|t|Т|т)(?:P|p|П|п)', ' ESTP ', post)
    post = re.sub ('(?:E|e|Е|е)(?:S|s|С|с)(?:T|t|Т|т)(?:J|j)', ' ESTJ ', post)
    post = re.sub ('(?:I|i|И|и)(?:N|n|Н|н)(?:F|f|Ф|ф)(?:P|p|П|п)', ' INFP ', post)
    post = re.sub ('(?:I|i|И|и)(?:N|n|Н|н)(?:F|f|Ф|ф)(?:J|j)', ' INFJ ', post)
    post = re.sub ('(?:I|i|И|и)(?:N|n|Н|н)(?:T|t|Т|т)(?:P|p|П|п)', ' INTP ', post)
    post = re.sub ('(?:I|i|И|и)(?:N|n|Н|н)(?:T|t|Т|т)(?:J|j)', ' INTJ ', post)
    post = re.sub ('(?:I|i|И|и)(?:S|s|С|с)(?:F|f|Ф|ф)(?:P|p|П|п)', ' ISFP ', post)
    post = re.sub ('(?:I|i|И|и)(?:S|s|С|с)(?:F|f|Ф|ф)(?:J|j)', ' ISFJ ', post)
    post = re.sub ('(?:I|i|И|и)(?:S|s|С|с)(?:T|t|Т|т)(?:P|p|П|п)', ' ISTP ', post)
    post = re.sub ('(?:I|i|И|и)(?:S|s|С|с)(?:T|t|Т|т)(?:J|j)', ' ISTJ ', post)
    post = re.sub ('(?:M|m|М|м)(?:B|b|Б|б)(?:T|t|Т|т)(?:I|i|И|и)', ' MBTI ', post)

    regex = '' + chr (8203)
    post = re.sub (regex, ' ', post) # remove zero-width spaces

    post = re.sub ('\n', 'ALRA', post)
    post = re.sub ('\s', '  ', post)
    post = re.sub ('ALRA', ' \n ', post)

    # misc punctuation 1
    post = re.sub ('\.{3,}', ' ellipsis ', post)
    post = re.sub ('…', ' ellipsis ', post)
    post = re.sub ('—|–|―', '-', post)
    post = re.sub (' -{2,} ', ' - ', post)
    post = re.sub (' _{2,} ', '_', post)
    post = re.sub (';{2,}', ';', post)
    #

    #post = re.sub (' - ', ' ', post)
    #post = re.sub (' [.,!?;:] ', ' ', post)

    post = replaceEmoticons (post)
    post = re.sub (':[a-zA-Zа-яА-ЯёЁ]+:', ' emoticon ', post)

    # misc punctuation 2
    post = re.sub ('!{2,}', '!', post)
    post = re.sub ('\?{2,}', '?', post)
    post = re.sub ('(?:!\?|\?!)+', '?!', post)
    post = re.sub ('!\.{3,}', '!..', post)
    post = re.sub ('\?\.{3,}', '?..', post)
    #

    post = re.sub ('\d?\d?\d\s?%', ' percentage ', post)
    post = re.sub ('\d?\d:\d\d?', ' timedesignation ', post)
    post = re.sub ('\d+[,/]\d+', ' fraction ', post)
    post = re.sub ('-?\d{2,}', ' multidigit ', post)
    post = re.sub ('\d\.?\)|\d\.', ' listitem ', post) # list, e.g. 1.)
    post = re.sub ('\d-\d[.,!?;:]?', ' numhyphennum ', post) # e.g. 1-2
    post = re.sub ('\s\d[\s.,!?;]', ' singledigit ', post)
    post = re.sub ('@[a-zA-Z]{2,}', ' mentionsmb ', post) # @nickname

    post = re.sub (' {2,}', ' ', post)
    post = re.sub ('(?:\n\s*\n)+', '\n', post)
    if (post [0] == ' '):
        post = post [1: ]

    return post
```

```python
parsed = []
for i in range (0, data1 ['without_url'].size):
    parsed.append (filterAll (data1.loc [i, 'without_url']))

data1 ['without_url'] = parsed
data1.head (10)
```

```python
lstEI = []; lstNS = []; lstFT = []; lstPJ = []
for i in range (0, data1 ['type'].size):
    lstEI.append (data1.loc [i, 'type'] [0] == 'I')
    lstNS.append (data1.loc [i, 'type'] [1] == 'S')
    lstFT.append (data1.loc [i, 'type'] [2] == 'T')
    lstPJ.append (data1.loc [i, 'type'] [3] == 'J')

data1 ['E/I'] = lstEI
data1 ['N/S'] = lstNS
data1 ['F/T'] = lstFT
data1 ['P/J'] = lstPJ

data1 = data1.drop (columns=['type'])
data1.head ()
```

```python
data1.to_csv ('without_url.csv', index_label = False)
```
